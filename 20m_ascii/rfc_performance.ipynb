```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, YearLocator
```

```python
#open csv
data = '/Users/meganmason491/Documents/research/sierra/analysis/results/tmp/SDV_featureImportance.csv'
```

```python
df = pd.read_csv(data, index_col='Date')
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
df.describe()
```

```python
ax = df.plot(marker='o', linestyle='', figsize=(20,8))

yloc = YearLocator(base=1, month=1, day=1)

ax.yaxis.set_major_locator(yloc)
plt.grid(axis='x', color='k', which='major', linestyle='-', linewidth=2)
ax.grid(True)
```
